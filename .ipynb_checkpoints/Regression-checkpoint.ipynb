### Suppression de colonnes 

```python
df_traite = df.drop(columns=['long','lat','zipcode','id','date','id','condition'])
```

```python
df_traite.columns
```

### Traiter le dataset (sup col date id )

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import pickle
```

```python
# DonnÃ©es d'exemple (X : variables indÃ©pendantes, Y : variable dÃ©pendante)
X = df_traite.drop(columns=['price'],axis=1) # Exemple de caractÃ©ristiques (par exemple, surface et nombre de chambres)
Y = df_traite['price']  # Prix de la maison (exemple)

# Split des donnÃ©es en un ensemble d'entraÃ®nement et un ensemble de test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
print(X_train.columns)
```

```python
df_traite.head()
```

```python
lm = LinearRegression()
lm.fit(X_train, Y_train)            # apprentissage
y_pred = lm.predict(X_test)         # prÃ©diction sur l'ensemble de test
```

```python
plt.figure(figsize=(6,6))
plt.scatter(Y_test, y_pred)
plt.plot([Y_test.min(),Y_test.max()],[Y_test.min(),Y_test.max()], color='red', linewidth=3)
plt.xlabel("Prix")
plt.ylabel("Prediction de prix")
plt.title("Prix reels vs predictions")
```
